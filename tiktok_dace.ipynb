```python
def get_unit_data(url):
    # fecha del scraping
    today = date.today()
    now = datetime.now()

    fecha_webscraping = today.strftime("%d/%m/%Y") + " " + now.strftime("%H:%M")
    
    
    
    try:
        website = browser.find_element_by_class_name('tiktok-847r2g-SpanLink')
        website = website.text
    except NoSuchElementException:
        website = "--"
        
    try:
        nombre_perfil = browser.find_element_by_class_name('tiktok-qpyus6-H1ShareSubTitle')
        nombre_perfil = nombre_perfil.text
    except NoSuchElementException:
        nombre_perfil = "--"
        
    link = 'https://www.tiktok.com/@'+ nombre_perfil
        
    try:
        desc_perfil = browser.find_element_by_class_name('tiktok-1n8z9r7-H2ShareDesc')
        desc_perfil = desc_perfil.text
    except NoSuchElementException:
        desc_perfil = "--"
    
    try:
        Following = browser.find_element_by_xpath('//*[@id="app"]/div[2]/div[2]/div/div[1]/h2[1]/div[1]/strong')
        Following = Following.text
    except NoSuchElementException:
        Following = "--"
        
    try:
        Followers = browser.find_element_by_xpath('//*[@id="app"]/div[2]/div[2]/div/div[1]/h2[1]/div[2]/strong')
        Followers = Followers.text
    except NoSuchElementException:
        Followers = "--"
        
    try:
        Likes = browser.find_element_by_xpath('//*[@id="app"]/div[2]/div[2]/div/div[1]/h2[1]/div[3]/strong')
        Likes = Likes.text
    except NoSuchElementException:
        Likes = "--"

    return {
        "link" : link,
        "website" : website,
        "nombre_perfil" : nombre_perfil,
        "desc_perfil" : desc_perfil,
        "Following": Following,
        "Followers": Followers,
        "Likes":Likes,
        "fecha_webscraping":fecha_webscraping
    } 
```

```python
def normalize_text(text):
    text = text.strip()
    text = text.lower()
    return text
def extract_cellphones(text):
    celulares = []
    #extract
    text = normalize_text(text)
    text = re.sub(r'[^0-9]{1}', ' ', text)
    cell_posible_array = re.findall(r'5119[0-9]{8}',text)
    if(cell_posible_array!=None):
        for cell in cell_posible_array:
            new_cell = cell[3:]
            celulares.append(new_cell)
    cell_posible_array = re.findall(r'519[0-9]{8}',text)
    if(cell_posible_array!=None):
        for cell in cell_posible_array:
            new_cell = cell[2:]
            celulares.append(new_cell)
    cell_posible_array = re.findall(r'9[0-9]{2}[\s]{1,3}[0-9]{3}[\s]{1,3}[0-9]{3}',text)
    if(cell_posible_array!=None):
        for cell in cell_posible_array:
            new_cell = cell[0:].replace(' ','')
            celulares.append(new_cell)
    cell_posible_array = re.findall(r'9[0-9]{8}',text)
    if(cell_posible_array!=None):
        for cell in cell_posible_array:
            new_cell = cell[0:].replace(' ','')
            celulares.append(new_cell)
    #eliminar duplicados        
    celulares = list(dict.fromkeys(celulares))        
    return celulares
```

```python
final['Telef_contacto'] = final.apply(lambda x: extract_cellphones(x['desc_perfil']) ,axis =1)
```
